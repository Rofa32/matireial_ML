```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"act9.txt", delimiter='\t')
```

```python
def visualize_clusters(data, labels):
    plt.scatter(data['width'], data['height'], c=labels)
    plt.xlabel('Width')
    plt.ylabel('Height')
    plt.title('Cluster Visualization')
    plt.show()
```

```python
def apply_kmeans(data, k):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    labels = kmeans.labels_
    return labels
```

```python
def calculate_silhouette_score(data, labels):
    score = silhouette_score(data, labels)
    return score
```

```python
def normalize_data(data):
    scaler = MinMaxScaler()
    normalized_data = scaler.fit_transform(data)
    return normalized_data
```

```python
labels = apply_kmeans(data[['width', 'height']], k=4)
visualize_clusters(data, labels)
```

```python
normalized_data = normalize_data(data[['width', 'height']])
normalized_labels = apply_kmeans(normalized_data, k=4)
visualize_clusters(data, normalized_labels)
```

```python
silhouette_score_non_normalized = calculate_silhouette_score(data[['width', 'height']], labels)
silhouette_score_normalized = calculate_silhouette_score(normalized_data, normalized_labels)
print("Silhouette score (non-normalized):", silhouette_score_non_normalized)
print("Silhouette score (normalized):", silhouette_score_normalized)
```
